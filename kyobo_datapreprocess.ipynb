```python
def test(s):
  hangul = re.compile('[^ ㄱ-ㅣ 가-힣+]')
  result = hangul.sub('', s)
  return(result)
```

```python
from konlpy.tag import Kkma, Okt
```

```python
okt = Okt()
```

```python
def tokenization(korean_reviews):
  review_final = []
  for lyric in korean_reviews:
    stem_pos_results = okt.pos(lyric, stem = True)

    lst = []
    for a, b in stem_pos_results:
      if b == 'Noun' or b == 'VerbPrefix' or b == 'Adjective':
        lst.append(a)
    review_final += lst
  return review_final
```

```python
final = []
```

```python
for n in tqdm(range(len(df['1']))):
  sentence = df['1'][n]
  sentences = sentence.splitlines()

  korean_reviews = []
  for sent in sentences:
    korean_review = test(sent)
    korean_review = korean_review.strip()
    if korean_review == '':
      continue
    korean_reviews.append(korean_review)

  make_set = set(korean_reviews)
  korean_reviews = list(make_set)
  final.append(tokenization(korean_reviews))
```
